### Gráficos criados pelo Python dentro do Power BI

1. Criar Ambiente Virtual com Python 3.6
2. Instalar todas as bibliotecas que vai querer usar no Power BI. Normalmente instalo de cara:
    - matplotlib
    - seaborn
    - pandas e openpýxl
    - jupyter
3. Sincronizar o Ambiente Virtual com o Power BI
4. Criar os gráficos

Obs: Sempre garanta que os gráficos que você quer criar funcionam no Python, porque o Power BI não exibe uma boa mensagem de erro

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_excel("Cotacoes.xlsx", engine='openpyxl')
display(dataset)
```

### Gráfico Normal

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(dataset["Date"], dataset["Adj Close"], "b--")
plt.show()
```

### Gráfico Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(dataset[["Date", "High", "Volume", "Adj Close"]])
```
